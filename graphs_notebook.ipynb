```python
# plot using sns
sns.set_theme(style="darkgrid")
# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(15, 10))
sns.despine(f, left=True, bottom=True)
g = sns.scatterplot(x="Vehicle Cost [$/vehicle]", y="Multi Attribute Utility",
                hue="Pack Type", style="Motor & Inverter",s=150, alpha=0.5,
                #palette="ch:r=-.2,d=.3_r",
                sizes=(20, 200), linewidth=0,
                legend="brief",
                data=df, ax=ax)
#g4.set_xlim(60000, 200000)
g.set_title("Vehicle Cost vs MAU")
g.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), ncol=1)
```

# Filtering data by requirements

Plan:
1. For task 3, create a subset that contains level 3 and above of autonomy. Plot the tradespace & explore for other insights
2. For task 4, create a subset that contains autonomy level 4 and above. Plot the tradespace & explore for other insights
3. For task 5, create a subset that contains autonomy level 5 and above. Plot the tradespace & explore for other insights

```python
# filtering for constraints and requirements 1-6
df_lvl0 = df[(df['C1']==1) & (df['C2']==1) &\
             (df["R1"]==1) & (df['R2']==1) &\
             (df["R3"]==1) & (df['R4']==1) &\
             (df["R5"]==1) & (df['R6']==1)]
```

```python
# sanity check
df_lvl0.info()
```

```python
# sanity check 2
df_lvl0.head(5)
```

```python
# plot using sns
sns.set_theme(style="darkgrid")
# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(15, 10))
sns.despine(f, left=True, bottom=True)
#clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
g = sns.scatterplot(x="Vehicle Cost [$/vehicle]", y="Multi Attribute Utility",
                hue="Pack Type", style="Motor & Inverter",
                #palette="ch:r=-.2,d=.3_r",
                sizes=(20, 200), linewidth=0,
                legend="brief",
                data=df_lvl0, ax=ax)
g.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), ncol=1)
```
